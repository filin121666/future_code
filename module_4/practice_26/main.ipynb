```python
import pandas as pd

df = pd.read_csv('car_data.csv')
cat = df.describe()

def choose_cat(price, one_cond, two_cond):
    if price < one_cond:
        return 'дешёвый'
    elif price > one_cond and price < two_cond:
        return 'нормальный'
    elif price > two_cond:
        return 'дорогой'

one_cond = cat['Present_Price']['25%']
two_cond = cat['Present_Price']['75%']

df['Car_Category'] = df['Present_Price'].apply(lambda x: choose_cat(x, one_cond, two_cond))

df
```
